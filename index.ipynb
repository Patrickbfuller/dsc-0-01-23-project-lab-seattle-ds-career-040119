```python
import string
(string.punctuation)
```

```python
bad_chars = list(string.punctuation.replace("'",""))
bad_chars
```

```python
bad_chars += ['\n', '\r']
bad_chars += string.digits
# bad_chars.append()
bad_chars[-15:]
```

```python
word = 'party!@#$'
for bad_char in bad_chars:
    word = word.replace(bad_char, '')
word
```

```python
words_dict = {}
words_dict.keys()
for word in split_test_text:
    
        #homogenize words that have different capilizations
    word = word.lower()
    
        #remove non_letters(but leave apostrophes for contractions)
    for bad_char in bad_chars:
        word = word.replace(bad_char, '')
    
    if word in words_dict.keys():
        words_dict[word] += 1
        continue
    words_dict.update({word : 1})
words_dict
```

```python
words_dict = {}
words_dict.keys()
for word in macbeth_play.split(' '):
    
        #homogenize words that have different capilizations
    word = word.lower()
    
        #remove non_letters(but leave apostrophes for contractions)
    for bad_char in bad_chars:
        word = word.replace(bad_char, '')
    
    if word in words_dict.keys():
        words_dict[word] += 1
        continue
    words_dict.update({word : 1})
words_dict
```

```python
len(words_dict.values())
```

```python
max(words_dict.values())
```

```python
# get top 25 
sorted(words_dict.values(), reverse=True)[:26]
```

```python
top_25_counts = sorted(words_dict.values(), reverse=True)[:26]
len(top_25_counts)
```

```python
top_words_dict = {}
for word in words_dict:
    if words_dict[word] in top_25_counts:
        top_words_dict.update({word : words_dict[word]})
top_words_dict.pop('')
len(top_words_dict), top_words_dict
```

```python
top_words_dict.items()
```

```python
top_words_sorted = sorted(top_words_dict.items(), key = lambda x: x[1], reverse = True)
top_words_sorted
```

```python
[x[1] for x in top_words_sorted]
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12,12))

# topwords = list(top_words_dict.keys())
# wordcounts = list(top_words_dict.values())

topwords = [x[0] for x in top_words_sorted]
topcounts = [x[1] for x in top_words_sorted]

plt.barh(topwords, topcounts)

plt.xlabel('Number of times word appears')
plt.ylabel('Most common words')

plt.title('The words appearing in Macbeth most frequently')
plt.show
```
